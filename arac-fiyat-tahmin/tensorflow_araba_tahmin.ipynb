```python
yeni_df = yeni_df[yeni_df.price != 1970]
```

```python
yeni_df.head()
```

```python
yeni_df.drop("transmission", axis=1, inplace=True)
```

```python
yeni_df.head()
```

```python
y = yeni_df["price"].values
X = yeni_df.drop("price", axis=1).values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

x_train[:5]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_train[:5]
```

```python
x_test[0:5]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x_train.shape
```

```python
model = Sequential()
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam" , loss="mse")

model.fit(x_train, y_train, epochs=300, validation_data=(x_test, y_test), batch_size=250)
```

```python
kayip_verisi = pd.DataFrame(model.history.history)
kayip_verisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

tahmin_dizisi = model.predict(x_test)
mean_absolute_error(y_test, tahmin_dizisi)
```
